```python
import xml.etree.ElementTree as ET
import pandas as pd
# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
# Load the XML file
tree = ET.parse('3.xml')
root = tree.getroot()

# Initialize an empty list to store the extracted data
data = []

# Iterate through all TRX_Sale elements
for trx_sale in root.findall(".//TRX_Sale"):
    order_element = trx_sale.find(".//Order")
    
    if order_element is not None:
        order_attributes = order_element.attrib
        data.append(order_attributes)

# Create a pandas DataFrame from the extracted data
df = pd.DataFrame(data)

# Display the DataFrame
display(df)
```

```python
import xml.etree.ElementTree as ET
import pandas as pd
# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
# Load the XML file
tree = ET.parse('3.xml')
root = tree.getroot()
# Initialize an empty list to store the extracted data
data = []

# Iterate through all TRX_Sale elements
for trx_sale in root.findall(".//TRX_Sale"):
    order_element = trx_sale.find(".//Order")
    
    if order_element is not None:
        order_attributes = order_element.attrib
        
        custom_info_element = order_element.find(".//CustomInfo")
        if custom_info_element is not None:
            vat_invoice_info = custom_info_element.find("./Info[@name='VATInvoiceNumber']")
            if vat_invoice_info is not None:
                order_attributes['VATInvoiceNumber'] = vat_invoice_info.attrib['value']
        
        data.append(order_attributes)

# Create a pandas DataFrame from the extracted data
df = pd.DataFrame(data)
# Display the DataFrame
display(df)
```

```python
import xml.etree.ElementTree as ET
import pandas as pd

# Load the XML file
tree = ET.parse('3.xml')
root = tree.getroot()

# Initialize an empty list to store the extracted tender data
tenders_data = []

# Iterate through all Tender elements
for tender in root.findall(".//Tender"):
    tender_data = {
        'TenderId': tender.findtext('TenderId').strip(),
        'TenderName': tender.findtext('TenderName').strip(),
        'TenderAmount': tender.findtext('TenderAmount').strip()
    }
    tenders_data.append(tender_data)

# Create a pandas DataFrame from the extracted tender data
tenders_df = pd.DataFrame(tenders_data)

# Display the tender DataFrame
display(tenders_df)
```

```python
import xml.etree.ElementTree as ET
import pandas as pd

# Load the XML file
tree = ET.parse('3.xml')
root = tree.getroot()

# Initialize lists to store extracted data
pos_list = []
pod_list = []
check_refund_list = []

# Find all TRX_BaseConfig elements
base_configs = root.findall(".//TRX_BaseConfig")

# Iterate through each TRX_BaseConfig element
for base_config in base_configs:
    pos = base_config.get('POS')
    pod = base_config.get('POD')
    check_refund = base_config.find('Config/CheckRefund').text.strip()
    
    # Append data to respective lists
    pos_list.append(pos)
    pod_list.append(pod)
    check_refund_list.append(check_refund)

# Create a pandas DataFrame
data = {'POS': pos_list, 'POD': pod_list, 'CheckRefund': check_refund_list}
base_config_df = pd.DataFrame(data)

# Display the base config DataFrame
display(base_config_df)
```
